```python
import numpy as np
from time import sleep
from IPython.display import clear_output
import matplotlib.pyplot as plt
import pickle
import matplotlib.pyplot as plt
```

```python
#setting up global variables
rows = 5
cols = 5
number_states = rows * cols
a = (0,1)
A = (4, 1)
b = (0, 3)
B = (2, 3)
start_state = (np.random.randint(0, rows), np.random.randint(0, cols))
n_trails = 100
prob = 0.2
gamma = .99

class gridworld:
    def __init__(self, position=start_state, n_trails=n_trails, display=False):
        
        self.display = display
        
        self.prob = prob
        self.gamma = gamma
        #duration
        self.n_trails = n_trails

        #creating the grid
        self.grid = np.zeros([rows, cols])
        
        #states 
        self.previous_states = [(0,0)]
        self.current_state = position
        
        #rewards
        self.reward = 0
        
        #tracking

        self.total_rewards = 0
        self.all_actions = []
        self.all_rewards = []
        self.all_states = []
        
    
    def move(self, action):
        """
        Moves the agent to a new location on the grid
        
        0 = up
        1 = down
        2 = left 
        3 = right
        """
        
        if action == 0:
            next_state = (self.current_state[0] - 1, self.current_state[1])
            
        elif action == 1:
            next_state = (self.current_state[0] + 1, self.current_state[1])
            
        elif action == 2:
            next_state = (self.current_state[0], self.current_state[1] - 1)
            
        elif action == 3:
            next_state = (self.current_state[0], self.current_state[1] + 1)
        
        if (next_state[0] >= 0) & (next_state[0] <= rows - 1):
            if (next_state[1] >= 0) & (next_state[1] <= cols - 1):
                return next_state
        
        return self.current_state
    
    def give_reward(self):
        """
        gives agent a reward or punishment based on its location on the grid
        """
        if (self.current_state == a):
            self.current_state = A
            return 10
        
        elif (self.current_state == b):
            self.current_state = B
            return 5
        
        elif (self.current_state == self.previous_states[-1]):
            return -1
        else:
            return 0
        
    def show(self):
        """
        Visualize the agents location on the grid
        """
        temp = np.zeros([rows, cols])
        temp[self.current_state] = 1
        print(temp)
        self.all_states.append(temp)
        
    def run(self, agent):
        """
        Plays the game
        
        requires:
        agent with choose & update functions
        """
        for i in range(self.n_trails):
            
            action = agent.choose()
            self.current_state = self.move(action)
            
            self.reward = self.give_reward()
            self.total_rewards += self.reward
            
            if self.display == True:
                clear_output(wait=True)
                self.show()
                print('old state:{}, action:{}, new state:{}, reward:{}'.format(self.previous_states[-1], action, self.current_state, self.reward))
                sleep(.5) 
            
            self.previous_states.append(self.current_state)
            self.all_actions.append(action)
            self.all_rewards.append(self.reward)
            
            agent.update()
            
            
        return self.total_rewards

        
```

```python
#just setting up some base things to track for the agents
class BaseSampliler:
    
    def __init__(self, env, table_size=None, learning_rate=None, epsilon=None, discount=None, table=False, theta=None, prob=None, gamma=None):

        self.env = env
        self.choice = 0
        self.state = env.current_state
        
        self.reward = env.reward
        
```

```python
#agent that selects moves at random
class random(BaseSampliler):
    def __init__(self, env):
        super().__init__(env)
        
    def choose(self):
        
        self.choice = np.random.randint(0,4)
        
        return self.choice
    def update(self):
        pass
```

```python
lookup_table = {}
for row in range(rows):
    for col in range(cols):
        if (row,col) == a:
            lookup_table[(row,col)] = 10
        elif (row,col) == b:
            lookup_table[(row,col)] = 5
        else:
            lookup_table[(row,col)] = 0
lookup_table

state_to_flat_convert = {}
flat_to_state_convert = {}
count = 0
for row in range(rows):
    for col in range(cols):
        state_to_flat_convert[(row, col)] = count
        flat_to_state_convert[count] = row, col
        count += 1

state = (1,1)
gamma = 0.001

def sprime(Values, action, state):
    
    new_state = state
    actions_dic = {
            0: (new_state[0] - 1, new_state[1]),
            1: (new_state[0] + 1, new_state[1]),
            2: (new_state[0], new_state[1] - 1),
            3: (new_state[0], new_state[1] + 1)
          }
    
    new_state = actions_dic[action]
    
    if new_state in lookup_table.keys():
        reward = lookup_table[new_state]
    else:
        new_state = state
        reward = -1
        
    state_prime = int(Values[state_to_flat_convert[new_state]])
    
    v = reward + gamma * state_prime - .1
    
    return v

Values = np.zeros(number_states)
Policy = np.zeros([number_states, 4])
force_stop = 1000

while True:
    stop = 0

    for state in range(25):
        all_actions = np.zeros(4)
        for action in range(4):
            all_actions[action] = sprime(Values, action, flat_to_state_convert[state])

            best_action_value = max(all_actions)
            best_action = np.argmax(all_actions)
            
            stop = max(stop, (best_action_value - Values[state]))
            
            Values[state] = best_action_value
            Policy[state] = np.eye(4)[best_action]
            
    if (stop < 0.001):
        break
        
    if force_stop == 0:
        break
    else:
        force_stop -= 1
        print(force_stop)

print(Values.reshape(5,5))
print()
print(Policy)

Policy_shaped = np.reshape(np.argmax(Policy, axis=1), [5,5])

class value_iteration(BaseSampliler):
    def __init__(self, env, policy):
        super().__init__(env, policy)
        self.policy = policy
        
    def choose(self):
        self.choice = self.policy[self.state]
        return self.choice
        
    def update(self):
        self.state = self.env.current_state
```

```python
#agent that selects moves using Q learning
class q_learn(BaseSampliler):
    def __init__(self, env, table_size, learning_rate, epsilon, discount, table=None):
        super().__init__(env, table_size, learning_rate, epsilon, discount, table=None)
        #exploration rate, will only explore for the first half of the trails
        self.epsilon = epsilon
        self.start_epsililon= 1 
        self.end_epsilon = self.env.n_trails // 2
        self.decay = epsilon/(self.end_epsilon - self.start_epsililon)
        
        #learning rate, how much each reward will update the Q_table
        self.lr = learning_rate
        self.discount = discount
        
        #setting lower and upper state bounds, not neccesary in a discrete state space like in this example
        self.low = np.array((0,0))
        self.high = np.array((4,4))
        
        #creating the qtable, uncomment the 3rd line if the table is not provided
        if table == False:
            self.table_size = table_size
            self.os_size = [self.table_size] * len(self.high)
            self.qtable = np.random.uniform(low=-1, high=0, size=(self.os_size + [4]))
        
        #supplying qtable, comment out if you dont provide or want to create a new one
        else:
            self.qtable = Q_table
        
        self.state
        
    def choose(self):
        '''
        makes a decision based on the highest value in the qtable or selects randomly if its an explore round
        '''
        if np.random.random() > self.epsilon:
            self.choice = np.argmax(self.qtable[self.state])
        else:
            self.choice = np.random.randint(0, 4)
    
        return self.choice
    
    def update(self):
        """
        updates the values to allow for learning
        """
        
        self.reward = self.env.reward
        
        self.current_q = self.qtable[self.state + (self.choice, )]
        
        self.state = self.env.current_state
        
        self.future_q = np.max(self.qtable[self.state])
        
        self.new_q = (1-self.lr) * self.current_q + self.lr * (self.reward+ self.discount * self.future_q)
        
        self.qtable[self.state + (self.choice, )] = self.new_q
        
        self.epsilon -= self.decay
```

```python
#random selection agent
r_fitness = []
for i in range(100):
    en0 = gridworld()
    r = random(en0)
    r_fitness.append(en0.run(agent=r))
    
    clear_output(wait=True)
    print(r_fitness[-1])
    plt.title('fitness')
    plt.axhline(0, c='r')
    plt.plot(range(len(r_fitness)), r_fitness)
    plt.show()
    sleep(0.1)
```

```python
#value iteration
vi_fitness = []
for i in range(100):
    en1 = gridworld()
    vi = value_iteration(en1, policy=Policy_shaped)
    vi_fitness.append(en1.run(vi))
    
    clear_output(wait=True)
    print(vi_fitness[-1])
    plt.title('fitness')
    plt.axhline(0, c='r')
    plt.plot(range(len(vi_fitness)), vi_fitness)
    plt.show()
    sleep(0.1)
```

```python
#q learning agent
q_fitness = []
for i in range(100):
    if i == 0:
        print('CREATING_NEW_TABLE')
        sleep(1)
        
        en2 = gridworld()
        q = q_learn(en2, table_size=20, learning_rate=.1, epsilon=.5, discount=.1, table=False)
        en2.run(agent=q)
        q_fitness.append(en2.run(agent=r))

        clear_output(wait=True)
        print(q_fitness[-1])
        plt.title('fitness')
        plt.axhline(0, c='r')
        plt.plot(range(len(q_fitness)), q_fitness)
        plt.show()
        sleep(0.25)
        Q_table = q.qtable
        
    else:
        en2 = gridworld()
        q = q_learn(en2, table_size=20, learning_rate=.1, epsilon=.3, discount=.1, table=True)
        en2.run(agent=q)
        q_fitness.append(en2.run(agent=r))

        clear_output(wait=True)
        print(q_fitness[-1])
        plt.title('fitness')
        plt.axhline(0, c='r')
        plt.plot(range(len(q_fitness)), q_fitness)
        plt.show()
        sleep(0.1)
        
#save the Qtable for later use
'''
Q_table = q.qtable
output = open('Q_Table.pkl', 'wb')
pickle.dump(Q_table, output)
output.close()
'''
```
